```python
# 모듈 호출
import pandas as pd
import numpy as np
import csv
```

```python
#data = pd.read_csv('./average_one.csv')
data = pd.read_csv('./average_group_correct.csv')
data.head()
```

```python
cols = list(data.columns)
df = data[cols[1:]]
```

```python
df.iloc[0, :7]
```

```python
# 분산 구하기
for i in range(0, len(df)):
    a= df.iloc[i, :7]*0.001
    v = np.var(a) # 분산
    print(i, data.iloc[i,0], v)
```

```python
result_csv = open('continue_rate_var_correct.csv', 'w', encoding='utf-8', newline='')
wr = csv.writer(result_csv)
wr.writerow(['name', 'mon_tue', 'tue_wed', 'wed_thu', 'thu_fri', 'fri_sat', 'sat_sun', 'sum', 'means', 'mm', 'var'])

for i in range(0, len(df)):
    name = data.iloc[i, 0]
    # 연속된 두 요일의 증감률
    array = []

    # 분산 구하기
    a = df.iloc[i, :7]*0.001
    v = np.var(a)
    
    for j in range(1, 7):
        prev = df.iloc[i, j-1] # j-1 : 0, 1, 2, 3, 4, 5
        next = df.iloc[i, j] # j : 1, 2, 3, 4, 5, 6
        array.append(((next - prev)/prev)*100)

    count = 0
    sum = 0
    for j in range(0,6):
        sum += array[j]
        if abs(array[j])>=300:
            count += 1
    means = sum/6
    
    min_num = min(df.iloc[i, :7])
    max_num = max(df.iloc[i, :7])
    mm = (max_num-min_num)*0.001
    if count == 0:
        wr.writerow([name, array[0], array[1], array[2], array[3], array[4], array[5], sum, means, mm, v])
        #print(i, name, array[0], array[1], array[2], array[3], array[4], array[5], v)
```

```python
result_csv.close()
```
